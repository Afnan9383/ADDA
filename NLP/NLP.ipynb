process this customer review using NLP techniques


```python
text='The phone is amazing and the camera quality is fantastic !'
```

Tokenization-split text into words

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

text='The phone is amazing and the camera quality is fantastic !'
tokens=word_tokenize(text)
print('tokens:',tokens)
```

stopwords removal -remove common words like 'is','and'...

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words=set(stopwords.words('english'))
filtered_tokens=[word for word in tokens if word.lower() not in stop_words]
print('after stopword removal:',filtered_tokens)
```

stemming-cut words to their root form

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed=[stemmer.stem(word) for word in filtered_tokens]
print('stemmed words:',stemmed)
```

lemmatization -convert word to dictionary form

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer=WordNetLemmatizer()
lemmatized=[lemmatizer.lemmatize(word) for word in filtered_tokens]
print('lemmatized words:',lemmatized)
```

POS Tagging- find if word is noun,verb,adjective...

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
pos_tags=nltk.pos_tag(filtered_tokens)
print('POS Tags:', pos_tags)
```

bag of words(BoW)-text to numeric frequency form

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus=[text]
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(corpus)

print('BoW words:', vectorizer.get_feature_names_out())
print('frequency matrix:',X.toarray())
```

TF-IDF - important words get higher scores

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(corpus)

print('TF-IDF Words:',vectorizer.get_feature_names_out())
print('TF-IDF score:',X.toarray())
```
