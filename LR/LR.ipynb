```python
#independent and dependent variable creation
x=data[['TV','Radio','Newspaper']]  # x variable contains all independent variables
y=data.Sales #y is output
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=99)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression() #object creation
```

```python
LR.fit(x_train,y_train) #training of linear regression
```

```python
y_predict=LR.predict(x_test) #test model
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error
```

```python
mean_squared_error(y_test,y_predict)
```

```python
r2_score(y_test,y_predict)
```
