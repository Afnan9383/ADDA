```python
from ultralytics import YOLO

#load YOLOv8 model
model = YOLO("yolov8n.pt")  #nano model (fast)

#run prediction on an image
results = model("dog_cat.jpg", conf=0.4)

# show results
results[0].show()
```

```python
from ultralytics import YOLO

#load YOLOv8 model
model = YOLO("yolov8m.pt")  #nano model (fast)

# detect image from coco dataset using class
results = model.predict(
    source = "horse.jpg",
    classes=[17], #horse
    conf=0.15,
    imgsz=640,
    save=True
)

results[0].show()
```
