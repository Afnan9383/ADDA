```python
#separate numerical and categorical columns
num_cols = ['ApplicantIncome',
            'CoapplicantIncome',
            'LoanAmount',
            'Loan_Amount_Term',
            'Credit_History']
cat_cols = ['Gender',
            'Married',
            'Dependents',
            'Education',
            'Self_Employed',
            'Property_Area',
            'Loan_Status']

```

```python
num_cols = ['ApplicantIncome',
            'CoapplicantIncome',
            'LoanAmount',
            'Loan_Amount_Term',
            'Credit_History']

for col in num_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
for col in num_cols:
    data[col].fillna(data[col].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
#outliers
plt.figure(figsize=(15,10), facecolor='white')

plotnumber = 1
for column in num_cols:
    plt.subplot(2,3,plotnumber)
    sns.boxplot(x=data[column])
    plt.title(column)
    plotnumber += 1

plt.tight_layout()
plt.show()

```

```python
num_cols = ['ApplicantIncome',
            'CoapplicantIncome',
            'LoanAmount',
            'Loan_Amount_Term',
            'Credit_History']

for col in num_cols:

    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Replace outliers with median
    data.loc[data[col] > upper_limit, col] = data[col].median()
    data.loc[data[col] < lower_limit, col] = data[col].median()
```

```python
plt.figure(figsize=(15,10))

plotnumber = 1
for column in num_cols:
    plt.subplot(2,3,plotnumber)
    sns.boxplot(x=data[column])
    plt.title(column)
    plotnumber += 1

plt.tight_layout()
plt.show()
```

Apply Log Transformation

```python
import numpy as np

data['ApplicantIncome'] = np.log1p(data['ApplicantIncome'])
data['CoapplicantIncome'] = np.log1p(data['CoapplicantIncome'])
data['LoanAmount'] = np.log1p(data['LoanAmount'])
```

```python
plt.figure(figsize=(15,10))

plotnumber = 1
for column in num_cols:
    plt.subplot(2,3,plotnumber)
    sns.boxplot(x=data[column])
    plt.title(column)
    plotnumber += 1

plt.tight_layout()
plt.show()
```

Apply IQR Again After Log

```python
cols = ['ApplicantIncome','LoanAmount']

for col in cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Capping instead of replacing
    data[col] = np.clip(data[col], lower_limit, upper_limit)
```

```python
plt.figure(figsize=(15,10))

plotnumber = 1
for column in num_cols:
    plt.subplot(2,3,plotnumber)
    sns.boxplot(x=data[column])
    plt.title(column)
    plotnumber += 1

plt.tight_layout()
plt.show()

```

Encode Binary Columns (Manual Mapping)

```python
data = pd.read_csv("loan_approved.csv")

data.columns = data.columns.str.strip()
data.rename(columns={'Loan_Status (Approved)':'Loan_Status'}, inplace=True)
```

```python
print(data.columns)
```

```python
data['Gender'] = data['Gender'].map({'Male':1, 'Female':0})
data['Married'] = data['Married'].map({'Yes':1, 'No':0})
data['Education'] = data['Education'].map({'Graduate':1, 'Not Graduate':0})
data['Self_Employed'] = data['Self_Employed'].map({'Yes':1, 'No':0})
data['Loan_Status'] = data['Loan_Status'].map({'Y':1,'N':0})
```

Encode Multi-Category Columns (One-Hot Encoding)

```python
data.info()
```

```python
data = pd.get_dummies(data, columns=['Dependents','Property_Area'], drop_first=True)
```

```python
data.info()
```

```python
bool_cols = data.select_dtypes(include=['bool']).columns
data[bool_cols] = data[bool_cols].astype(int)
```
