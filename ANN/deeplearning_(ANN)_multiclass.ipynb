```python
import pandas as pd
```

```python
df=pd.read_csv("iris_dataset.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["species"] = le.fit_transform(df["species"])
df
```

```python
x=df.iloc[:,:4]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#model building
ann=Sequential()
#add layers
ann.add(Dense(4,activation="relu",input_shape=(x_train.shape[1],)))
ann.add(Dense(10,activation="relu"))
ann.add(Dense(10,activation="relu"))
ann.add(Dense(3,activation="softmax"))
```

```python
ann.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
x_train.shape
```

```python
ann.fit(x_train,y_train,batch_size=210,epochs=50)
```

```python
loss,accuracy=ann.evaluate(x_test,y_test)

print("accuracy",accuracy)
print("loss",loss)
```

```python
import pickle
pickle.dump(ann,open("model.save","wb"))
```

```python
pickle.dump(scaler,open("scaler.save","wb"))
```

```python
#example prediction
import numpy as np
import pickle

# Reload the scaler object from the saved file
scaler = pickle.load(open("scaler.save","rb"))

sample=np.array([[5.1,3.5,1.4,0.2]])
sample_scaled = scaler.transform(sample)
pred_class =ann.predict(sample_scaled)
pred_label = df.species[pred_class.argmax()]
print("predicted class:",pred_label)
```
