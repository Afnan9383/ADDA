```python
#check outlier
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1

for column in data:
    if plotnumber<=9:
        ax=plt.subplot(12,3,plotnumber)
        sns.boxplot(data[column])
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Count',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
# List of columns (excluding 'Age') to process
clinical_f = [col for col in data.columns if col != "Age"]

# Outlier Detection and Handling
for x in clinical_f:
    # Step 1.1: Detect outliers outside the range [0, 3] for each column
    outliers = data[(data[x] < 0) | (data[x] > 3)]

    #Print outliers' indices if they exist
    if not outliers.empty:
        print(f"Outliers in {x}: {outliers.index.tolist()}")
```

```python
#outliers by capping values between 1 and 3 for each column
if x in clinical_f:
    data[x] = data[x].apply(lambda x: max(1, min(3, x)))

# Step 2: Print the updated DataFrame
print(data)
```

```python
X = data.drop('class', axis=1)
y = data['class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

Train Multiclass Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=5000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
y_prob = model.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(y_test, y_prob, multi_class='ovr')

print("ROC AUC Score:", roc_auc)
```

```python
print("Training Accuracy:", model.score(X_train, y_train))
```
